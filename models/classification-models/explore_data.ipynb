```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")
```

```python
# Read the target_type.csv file
df = pd.read_csv('../../data/target_type.csv')

print(f"Dataset shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
df.head()
```

```python
# Identify the binary target columns (excluding text columns)
target_columns = [col for col in df.columns if col not in ['first_target', 'second_target', 'incident_summary', 'first_civilian_target', 'second_civilian_target']]

print("Target columns to analyze:")
for col in target_columns:
    print(f"  - {col}")

print(f"\nTotal target columns: {len(target_columns)}")
```

```python
# Count occurrences of each target type using .sum()
target_counts = df[target_columns].sum().sort_values(ascending=False)

print("Target Type Distribution:")
print("=" * 50)
for target, count in target_counts.items():
    percentage = (count / len(df)) * 100
    print(f"{target:25} | {count:5d} ({percentage:5.1f}%)")

print(f"\nTotal incidents: {len(df)}")
print(f"Total target labels: {target_counts.sum()}")
```

```python
# Create a bar plot to visualize the distribution
plt.figure(figsize=(12, 8))
bars = plt.bar(range(len(target_counts)), target_counts.values, color='skyblue', edgecolor='navy', alpha=0.7)

# Customize the plot
plt.title('Distribution of Target Types in SATP Dataset', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Target Type', fontsize=12)
plt.ylabel('Number of Incidents', fontsize=12)
plt.xticks(range(len(target_counts)), target_counts.index, rotation=45, ha='right')

# Add value labels on top of bars
for i, (bar, count) in enumerate(zip(bars, target_counts.values)):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10, 
             str(count), ha='center', va='bottom', fontweight='bold')

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Identify rare classes (less than 1% of total incidents)
rare_threshold = len(df) * 0.01
rare_classes = target_counts[target_counts < rare_threshold]

print("Rare Target Types (< 1% of incidents):")
print("=" * 40)
for target, count in rare_classes.items():
    percentage = (count / len(df)) * 100
    print(f"{target:25} | {count:3d} ({percentage:4.1f}%)")

print(f"\nNumber of rare classes: {len(rare_classes)}")
print(f"Rare threshold: {rare_threshold:.0f} incidents")
```

```python
# Summary statistics
print("Dataset Summary:")
print("=" * 30)
print(f"Total incidents: {len(df):,}")
print(f"Total target types: {len(target_columns)}")
print(f"Most common target: {target_counts.index[0]} ({target_counts.iloc[0]} incidents)")
print(f"Least common target: {target_counts.index[-1]} ({target_counts.iloc[-1]} incidents)")
print(f"Average incidents per target: {target_counts.mean():.1f}")
print(f"Median incidents per target: {target_counts.median():.1f}")

# Class imbalance ratio
max_count = target_counts.max()
min_count = target_counts.min()
imbalance_ratio = max_count / min_count if min_count > 0 else float('inf')
print(f"Class imbalance ratio (max/min): {imbalance_ratio:.1f}:1")
```
