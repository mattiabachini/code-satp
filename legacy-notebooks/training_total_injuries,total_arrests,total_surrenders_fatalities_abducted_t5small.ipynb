```python
import pandas as pd

def bin_count(x):
   if x == 0:
        return 'zero'
   else:
        return 'non-zero'

# Apply binning to each target variable
data['injuries_bin'] = data['total_injuries'].apply(bin_count)
data['arrests_bin'] = data['total_arrests'].apply(bin_count)
data['surrenders_bin'] = data['total_surrenders'].apply(bin_count)
data['fatalities_bin'] = data['total_fatalities'].apply(bin_count)
data['abducted_bin'] = data['total_abducted'].apply(bin_count)

# Create a combined stratification label
data['stratify_label'] = (
    data['injuries_bin'] + '_' +
    data['arrests_bin'] + '_' +
    data['surrenders_bin'] + '_' +
    data['fatalities_bin'] + '_' +
    data['abducted_bin']
)
```

```python
# Initialize empty lists for indices
train_indices = []
val_indices = []
test_indices = []

# Group data by the stratification label
grouped = data.groupby('stratify_label')

for _, group in grouped:
    n_samples = len(group)

    if n_samples >= 3:
        # Determine sizes
        train_size = int(0.6 * n_samples)
        val_size = int(0.2 * n_samples)

        # Adjust sizes if necessary
        remainder = n_samples - (train_size + val_size)
        test_size = n_samples - train_size - val_size

        # Ensure at least one sample per set
        if train_size == 0:
            train_size = 1
        if val_size == 0 and n_samples - train_size > 1:
            val_size = 1
        if test_size == 0 and n_samples - train_size - val_size > 0:
            test_size = 1

        # Shuffle the group indices
        group = group.sample(frac=1, random_state=42)

        # Split indices
        train_idx = group.iloc[:train_size].index
        val_idx = group.iloc[train_size:train_size+val_size].index
        test_idx = group.iloc[train_size+val_size:].index

    elif n_samples == 2:
        # Assign one sample to train and one to test
        train_idx = group.sample(n=1, random_state=42).index
        remaining = group.drop(train_idx)
        val_idx = remaining.sample(n=0).index  # Empty
        test_idx = remaining.index
    else:
        # If only one sample, assign it to training set
        train_idx = group.index
        val_idx = []
        test_idx = []

    # Add indices to the respective lists
    train_indices.extend(train_idx)
    val_indices.extend(val_idx)
    test_indices.extend(test_idx)
```

```python

# Create dataframes for each set
train_data = data.loc[train_indices]
val_data = data.loc[val_indices]
test_data = data.loc[test_indices]

# Reset indices (optional)
train_data = train_data.reset_index(drop=True)
val_data = val_data.reset_index(drop=True)
test_data = test_data.reset_index(drop=True)

# Verify the splits
print(f"Total samples: {len(data)}")
print(f"Training samples: {len(train_data)}")
print(f"Validation samples: {len(val_data)}")
print(f"Test samples: {len(test_data)}")
print(f"Sum of splits: {len(train_data) + len(val_data) + len(test_data)}")
```

```python
def create_input_output_pairs(row):
    context = row['incident_summary']
    qa_pairs = []

    # Injuries
    qa_pairs.append({
        'input_text': f"question: How many injuries occurred in the incident? context: {context}",
        'target_text': str(f"{int(row['total_injuries'])} injuries")
    })

    # Arrests
    qa_pairs.append({
        'input_text': f"question: How many arrests were made in the incident? context: {context}",
        'target_text': str(f"{int(row['total_arrests'])} arrests")
    })

    # Surrenders
    qa_pairs.append({
        'input_text': f"question: How many people surrendered in the incident? context: {context}",
        'target_text': str(f"{int(row['total_surrenders'])} surrenders")
    })

    # Fatalities
    qa_pairs.append({
        'input_text': f"question: How many fatalities occurred in the incident? context: {context}",
        'target_text': f"{int(row['total_fatalities'])} fatalities"
    })

    # Abducted
    qa_pairs.append({
        'input_text': f"question: How many people were abducted in the incident? context: {context}",
        'target_text': f"{int(row['total_abducted'])} abducted"
    })

    return qa_pairs
```

```python
# Training set
train_input_output_pairs = []
for idx, row in train_data.iterrows():
    qa_pairs = create_input_output_pairs(row)
    train_input_output_pairs.extend(qa_pairs)

# Validation set
val_input_output_pairs = []
for idx, row in val_data.iterrows():
    qa_pairs = create_input_output_pairs(row)
    val_input_output_pairs.extend(qa_pairs)

# Test set
test_input_output_pairs = []
for idx, row in test_data.iterrows():
    qa_pairs = create_input_output_pairs(row)
    test_input_output_pairs.extend(qa_pairs)
```

```python
# Create DataFrames
train_df = pd.DataFrame(train_input_output_pairs)
val_df = pd.DataFrame(val_input_output_pairs)
test_df = pd.DataFrame(test_input_output_pairs)
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
from transformers import Seq2SeqTrainer, Seq2SeqTrainingArguments
training_args = Seq2SeqTrainingArguments(
    output_dir='./t5_results',
    evaluation_strategy='epoch',
    learning_rate=3e-4,
    per_device_train_batch_size=48,
    per_device_eval_batch_size=48,
    num_train_epochs=1,
    weight_decay=0.01,
    save_total_limit=2,
    predict_with_generate=True,
    fp16=torch.cuda.is_available(),
)
```

```python
# Define compute_metrics if not already defined
import re
from sklearn.metrics import mean_absolute_error

def extract_number(text):
    # Use regex to find the first integer in the text
    match = re.search(r'\b\d+\b', text)
    if match:
        return int(match.group())
    else:
        return 0  # Default to 0 if no number is found

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Compute Exact Match
    exact_matches = [int(pred.strip() == label.strip()) for pred, label in zip(decoded_preds, decoded_labels)]
    em = sum(exact_matches) / len(exact_matches) * 100

    # Extract numbers from the predictions and labels
    pred_numbers = [extract_number(pred) for pred in decoded_preds]
    label_numbers = [extract_number(label) for label in decoded_labels]

    # Compute MAE using scikit-learn
    mae = mean_absolute_error(label_numbers, pred_numbers)

    return {'exact_match': em, 'mae': mae}
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python

trainer.train()
```

```python
# Evaluate the model
eval_results = trainer.evaluate()

print(f"Exact Match: {eval_results['eval_exact_match']:.2f}")
print(f"Mean Absolute Error: {eval_results['eval_mae']:.2f}")
```

```python
import torch
import re

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

def extract_number(text):
    # Use regex to find the first integer in the text
    match = re.search(r'\b\d+\b', text)
    if match:
        return int(match.group())
    else:
        return 0  # Default to 0 if no number is found

def predict_counts(incident_summary):
    questions = [
        ("How many injuries occurred in the incident?", "injuries"),
        ("How many arrests were made in the incident?", "arrests"),
        ("How many people surrendered in the incident?", "surrenders"),
        ("How many fatalities occurred in the incident?", "fatalities"),
        ("How many people were abducted in the incident?", "abducted")
    ]
    counts = {}
    for question, label in questions:
        input_text = f"question: {question} context: {incident_summary}"
        # Tokenize the input text
        input_ids = tokenizer.encode(input_text, return_tensors='pt', max_length=512, truncation=True)
        input_ids = input_ids.to(device)
        # Generate the output
        outputs = model.generate(input_ids)
        # Decode the output
        answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
        # Extract the numerical count
        count = extract_number(answer)
        counts[label] = count
    return counts
```

```python


# Separate data based on question type (injuries, arrests, surrenders)
injuries_df = df[df['true_label'].str.contains('injuries')]
arrests_df = df[df['true_label'].str.contains('arrests')]
surrenders_df = df[df['true_label'].str.contains('surrenders')]
fatalities_df = df[df['true_label'].str.contains('fatalities')]
abducted_df = df[df['true_label'].str.contains('abducted')]

# Accuracy for zero and non-zero true values for each category
def calculate_accuracy(df, label):
    zero_true = df[df['true_count'] == 0]
    non_zero_true = df[df['true_count'] != 0]

    zero_true_accuracy = (zero_true['true_count'] == zero_true['predicted_count']).mean()
    non_zero_true_accuracy = (non_zero_true['true_count'] == non_zero_true['predicted_count']).mean()

    return {
        f"{label} Zero True Accuracy": zero_true_accuracy * 100,
        f"{label} Non-Zero True Accuracy": non_zero_true_accuracy * 100
    }

# Calculate accuracies
injuries_accuracy = calculate_accuracy(injuries_df, "Injuries")
arrests_accuracy = calculate_accuracy(arrests_df, "Arrests")
surrenders_accuracy = calculate_accuracy(surrenders_df, "Surrenders")
fatalities_accuracy = calculate_accuracy(fatalities_df, "Fatalities")
abducted_accuracy = calculate_accuracy(abducted_df, "Abducted")


# Combine results
accuracy_results = {**injuries_accuracy, **arrests_accuracy, **surrenders_accuracy, **fatalities_accuracy, **abducted_accuracy}

accuracy_results
```

```python
# prompt: inference code funtion that takes the argument of a summary and returns the counts of each one, one single funtion

import torch
import re
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load the tokenizer and model from the saved directory
tokenizer = T5Tokenizer.from_pretrained('/content/drive/MyDrive/SATP_data/total_injuries-arrests-surrenders-fatalities-abducted/t5small_finetuned_model')
model = T5ForConditionalGeneration.from_pretrained('/content/drive/MyDrive/SATP_data/total_injuries-arrests-surrenders-fatalities-abducted/t5small_finetuned_model')

# Set the model to evaluation mode
model.eval()

# Move the model to the appropriate device (GPU if available, else CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

def extract_number(text):
    match = re.search(r'\b\d+\b', text)
    if match:
        return int(match.group())
    else:
        return 0

def predict_counts(incident_summary):
    questions = [
        ("How many injuries occurred in the incident?", "injuries"),
        ("How many arrests were made in the incident?", "arrests"),
        ("How many people surrendered in the incident?", "surrenders"),
        ("How many fatalities occurred in the incident?", "fatalities"),
        ("How many people were abducted in the incident?", "abducted")
    ]
    counts = {}
    for question, label in questions:
        input_text = f"question: {question} context: {incident_summary}"
        input_ids = tokenizer.encode(input_text, return_tensors='pt', max_length=512, truncation=True)
        input_ids = input_ids.to(device)
        outputs = model.generate(input_ids)
        answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
        count = extract_number(answer)
        counts[label] = count
    return counts
```
